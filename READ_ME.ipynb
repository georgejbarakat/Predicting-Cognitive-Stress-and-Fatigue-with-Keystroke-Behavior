# Predicting Cognitive Stress and Fatigue with Keystroke Behavior

## Abstract

This project investigates the use of keystroke dynamics for predicting cognitive stress and mental fatigue using classical machine learning techniques. Behavioral features capturing typing speed, inter-key timing, rhythm variability, and error correction patterns are extracted from raw keystroke time-series data. We use two datasets: the large-scale IKDD keystroke dynamics dataset for feature engineering, statistical analysis, and behavioral modeling, and a labeled stress detection dataset from Kaggle for supervised training and performance evaluation. To address limited labeled sample sizes, we apply data augmentation via sliding window segmentation on time-series data to generate robust behavioral samples. Supervised models under consideration include logistic regression, random forests, support vector machines, and gradient boosted trees. The project culminates in a Streamlit web application enabling real-time stress and fatigue estimation from live user typing input.

## Methodology

Our pipeline begins with raw keystroke event logs from the IKDD dataset, from which we engineer session-level behavioral features such as mean and variance of inter-key intervals (dwell time, flight time), typing speed (characters per second), backspace/error correction rate, and rhythm variability (coefficient of variation of inter-key timing). These features are designed to serve as behavioral proxies for cognitive load and fatigue. The labeled Kaggle stress dataset provides ground-truth stress labels linked to keystroke and mouse behavioral signals, which we use to train and evaluate supervised classifiers. To augment the limited number of labeled user sessions, we apply a sliding window approach to generate multiple overlapping samples per session. Our baseline model is logistic regression trained on the engineered feature set. Statistical analysis includes exploratory data analysis (EDA), distribution fitting, hypothesis testing (e.g., comparing feature distributions across stress levels), and bootstrapping to estimate model performance confidence intervals. By Milestone 2, we plan to improve upon the baseline with ensemble methods (Random Forest, Gradient Boosted Trees) and SVM, and have development of our real-time stream-lit tool
